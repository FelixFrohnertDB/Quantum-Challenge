# Introduction

In "Task 2" a generative model using a variational circuit is explored. 
Initially, we start with a set of 4 random 4-qubit quantum states. The goal is then to create and train a variational circuit that transforms input states into the following predefined output states:
- If random state 1 is provided, it returns state |0011>
- If random state 2 is provided, it returns state |0101>
- If random state 3 is provided, it returns state |1010>
- If random state 4 is provided, it returns state |1100>

I chose this assignment because one of my main interests is in the area of variational quantum algorithms. My goal was to apply my knowledge from previous projects to a new problem and learn something new.

To solve this problem, I implement a simplified version of the filtering variational quantum eigensolver.https://arxiv.org/pdf/2106.10055.pdf 
I was interested in learning more about this algorithm and took this as an opportunity to deepen my understanding of its application to a combinatorial optimization problem. Furthermore, I worked with David Amaro and Matthias Rosenkranz, the authors of the F-VQE paper, on an earlier project which helped me understand the algorithm.

The main idea here is that applying a filtering operator $f=H^{-\tau}$ to a quantum state $|\Psi>$ shifts its probability distribution of eigenvectors in such a way that low energy states dominate. This idea borrows from imaginary time evolution. 

In general, any arbitrary wavefunction can be written as a superposition of eigenstates:


$$\psi(x,0) = \sum_i a_i(0) \psi_i(x) $$
and
$$ \psi(x,t) = \sum_i exp(-it E_n / \hbar ) a_i(0) \psi_i(x) $$


Eigenstate don't change in their overall magnitude with increasing time evolution but rotate with a frequency proportional to the corresponding energy eigenvalue $En/\hbar$.




With the substitution $(it \rightarrow{} \tau )$ 


$$ \psi(x,\tau) = \sum_i exp(- \tau E_n / \hbar ) a_i(0) \psi_i(x) $$

we now see that $\psi(x,t)$ at imaginary time no longer consists of an "oscillating" superposition of energy eigenstates, but of an "exponentially decaying" superposition of energy eigenstates. Hereby, the higher energy states are decaying faster, so that with time the state with the lowest energy dominates the wave function.


The F-VQE algorithm can then be summarized as follows:
- Prepare an initial state $|\psi_0>$ that has finite overlap with the ground state $P(\lambda_0) > 0$.
- Iteratively, the application of the filter operator $F_t$ to $|\psi_{t-1}>$ is approximated by a state $|\psi_t>$. 
For this purpose, a parameterized quantum circuit ansatz $|\psi(\theta)>$ is used, which depends on a vector of m parameters
$\theta = (\theta_1, . . . , \theta_m)$. 
At each optimization step t, the parameters that minimize the Euclidean distance between the parameterized quantum state and $|F_t \psi_{t-1}>$ are sought. For this purpose, a gradient-based approach is chosen, where 

$$\theta_t \propto \theta_{t-1}-\sum_i \frac{\partial C_t(\theta)}{\partial \theta_i} \propto \theta_{t-1}-\sum_i <\psi_{t-1}|F_t|\psi(\theta \pm \frac{\pi}{2})>$$
Simply put, we rotate the parameter values by $\pm \frac{\pi}{2}$, calculate each result, and choose which rotation works best.
- Repeat for a given number of optimization steps.

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from pytket.circuit import Circuit
from pytket.extensions.qiskit import AerStateBackend, AerBackend  
from pytket.circuit.display import render_circuit_jupyter
from pytket.utils.results import probs_from_state  
from qiskit.quantum_info import partial_trace, entropy, Statevector  
```

```python
def initial_ansatz(params):
    params = params / np.pi  # pytket works in units of pi
    param_index = 0
    c = Circuit(n_qubits)

    for q in range(n_qubits):
        c.Rx(params[param_index], q)
        param_index += 1
    for q in range(n_qubits):
        c.Ry(params[param_index], q)
        param_index += 1
    for q in range(n_qubits):
        c.Rz(params[param_index], q)
        param_index += 1

    # This creates a so called "Checkerboard"-Ansatz with the Cnot gates.
    # Usually, I get better performance compared to a "Ladder"-Ansatz, hence this choice
    for q in np.arange(0, n_qubits - 1, 2):
        c.CX(q, q + 1)
    for q in np.arange(1, n_qubits - 1, 2):
        c.CX(q, q + 1)
    return c

n_qubits = 4
n_gates = 3

#Initialize #Rx,Ry,Rz gates with normally distributed values of rotations.
initial_params = np.random.rand(n_gates*n_qubits)*np.pi/2
```

# Initialize full circuit


Now that we have a method to initialize a random state with a given set of initial parameters, we can build the rest of the circuit. Here, the structure of the gates $R_x$, $R_y$, $R_z$ and $Cnot$ is repeated in several layers.

```python
def ansatz(params):
    params = params / np.pi    
    param_index = 0
    c = Circuit(n_qubits)
    
    #Initialize one of four 'random' states
    for q in range(n_qubits):
        c.Rx(params[param_index], q)
        param_index += 1
    for q in range(n_qubits):
        c.Ry(params[param_index], q)
        param_index += 1
    for q in range(n_qubits):
        c.Rz(params[param_index], q)
        param_index += 1

    for q in np.arange(0, n_qubits - 1, 2):
        c.CX(q, q + 1)
    for q in np.arange(1, n_qubits - 1, 2):
        c.CX(q, q + 1)
        
    #Initialize rest of circuit used to solve the task
        
    for l in range(n_layers):
        for q in range(n_qubits):
            c.Rx(params[param_index], q)
            param_index += 1
        for q in range(n_qubits):
            c.Ry(params[param_index], q)
            param_index += 1
        for q in range(n_qubits):
            c.Rz(params[param_index], q)
            param_index += 1
            
        for q in np.arange(0, n_qubits - 1, 2):
            c.CX(q, q + 1)
        for q in np.arange(1, n_qubits - 1, 2):
            c.CX(q, q + 1)

    return c

n_qubits = 4
n_layers = 2 
n_gates = 3
n_params = n_qubits * n_layers *n_gates   


initial_params = r_state_1
circuit_params = np.random.rand(n_params)*np.pi/2
params = np.concatenate((initial_params,circuit_params))
```

# Defining cost function

A key aspect of any VQE algorithm is, of course, the cost function that we wish to minimize.
In our case, the structure of $C$ is simple. We want to penalize configurations that have the wrong bitsring value in the wrong place. For example, if the optimal solution is $(0,0,1,1)$, a configuration like $(1,1,0,0)$ should be assigned a high cost. This is realized by the following cost function:   

```python
def C1(x):
    solution = (0,0,1,1)
    return np.sum(np.abs(x-np.array(solution)))
```

# Defining F-VQE prerequisites

This section defines the auxiliary functions of the F-VQE "protocol". I have tried to arrange them in descending order of importance.
As mentioned earlier, an important component of the F-VQE is the filtering operator $f$. It is defined as a real function $f(c)$, with $c$ as a real parameter, where $f^2(c)$ is strictly decreasing. Furthermore, the $\tau>0$ function parameter can be varied at each optimization step.
Here, an inverse filter was chosen because it showed the best performance in the F-VQE study.

```python
def f(c, t):
    if c==0:
        return (c+0.0000001) ** (-t)    
    return c ** (-t)
```

The $\tau$ parameter is continuously adjusted to keep the gradient norm as close as possible to a desired large and fixed value to prevent the gradient from disappearing. 
 
In this case, the implementation is done using a simple heuristic diretly taken from the F-VQE paper.
At each iteration step, we want to choose $\tau$ such that the gradient norm is as close as possible but below a certain threshold $r$, i.e., we solve the implicit equation $\text{gradient_norm}(\tau) = r$


For this we evaluate the gradient norm for increasing values
of $\tau$ until an upper bound is found such that $\text{gradient_norm}(t_{up}) >= r$.
Then we search for $\tau$ in the range $[0, \tau_{up}]$ up to
a certain precision. 

```python
# Gradient where probs_plus/minus represents theta shifted by +- pi/2
# The partial derivative of a parameter is the difference 
# between two circuits that differ in a positive/negative pi/2 
# shift of the parameter. 
def g(probs_plus_all, probs_minus_all, t, spectrum):  
    output = np.zeros(n_params)
    for i in range(n_params):
        filter_exp_plus = evaluate_filter(probs_plus_all[i], t, spectrum)
        filter_exp_minus = evaluate_filter(probs_minus_all[i], t, spectrum)
        output[i] = filter_exp_minus - filter_exp_plus
    return output

def t_upper_bound(probs_plus_all, probs_minus_all, spectrum, n_trials=5):
    for t in (3**i for i in range(n_trials)):
        grad_norm = np.linalg.norm(g(probs_plus_all, probs_minus_all, t, spectrum))
        if grad_norm > r:
            return t, grad_norm
    print('Upper bound of t not found. Returning the largest value tested.' + \
          'Increase the number of trials or reduce the threshold for improvement')
    return t, grad_norm

# Once the upper bound is found, we look for the value of t in the range[0, t_up] that is closest and below r
def update_tau(probs_plus_all, probs_minus_all, spectrum):
    precision=r/10
    t_low = 0
    t_up, grad_norm = t_upper_bound(probs_plus_all, probs_minus_all, spectrum)
    while True:
        t = (t_up + t_low) / 2
        grad = g(probs_plus_all, probs_minus_all, t, spectrum)
        grad_norm = np.linalg.norm(grad)
        if r < grad_norm:
            t_up = t
        elif grad_norm < r - precision:
            t_low = t
        else:
            return t, grad
```

Samples from the two circuits for each parameter (differing in a positive/negative pi/2).

```python
def sample_shifted_ansatz(params):  
    probs_plus_all = {}
    probs_minus_all = {}
    for i in range(n_params):
        params_plus = params + np.bincount([i], None, n_params) * np.pi / 2
        params_minus = params - np.bincount([i], None, n_params) * np.pi / 2
        
        probs_plus = probs_from_state(get_state(params_plus))
        probs_minus = probs_from_state(get_state(params_minus))

        probs_plus_all[i] = probs_plus
        probs_minus_all[i] = probs_minus
    return probs_plus_all, probs_minus_all
```

```python
def sample_ansatz(params, state):
    c = ansatz(params)
    probs = probs_from_state(state)
    return probs

def get_state(params):
    c = ansatz(params)
    b = AerStateBackend()
    return b.get_state(c)
```

Function that updates the spectrum dictionary every time we sample a quantum circuit. The best solution will be the bitstring with the smallest cost that we have ever sampled.

```python
def update_spectrum(probs, spectrum,cost_function):
    for bitstring, p in probs.items():
        if bitstring in spectrum:
            cost = spectrum[bitstring]
        else:
            cost = cost_function(bitstring)
            spectrum[bitstring] = cost
    return spectrum
```

Gradient-based optimiser

```python
def update_params(params, gradient):
    return params - learning_rate * gradient
```

Evaluate the filter operator from the samples of a quantum circuit.  

```python
lower_bound = 0 # min cost
upper_bound = 4 # max cost
def rescale(c, lower_bound=lower_bound, upper_bound=upper_bound):
    b = 1 / (upper_bound - lower_bound)
    a = - lower_bound / (upper_bound - lower_bound)
    return a + b * c

def evaluate_filter(probs, t, spectrum):
    output = 0
    for bitstring, p in probs.items():
        cost = spectrum[bitstring]
        cost = rescale(cost)
        output += f(cost, t) * p
    return output
```

Function evaluating the average of the cost function with a given set of state probabilities.

```python
def evaluate_cost(probs, spectrum):
    output = 0
    for bitstring, p in probs.items():
        cost = spectrum[bitstring]
        output += cost * p
    return output
```

There are a number of analyses that can be done with VQEs. I have decided to limit myself to one property of interest, namely entaglement. Entaglement plays an important role in the efficiency of our variational algorithms, as described in https://arxiv.org/pdf/2103.14479.pdf.
The implementation of this entanglement measure is simply adapted from https://journals.aps.org/pra/abstract/10.1103/PhysRevA.76.022328.

```python
def get_entanglement(state):
    Q = 0
    for i in range(n_qubits):
        trace_over = [q for q in range(n_qubits) if q != i]  
        reduced_density_matrix = partial_trace(state, trace_over).data  
        #raise matrix to power 2 and trace it
        Q += np.real((np.linalg.matrix_power(reduced_density_matrix, 2)).trace())
    Q = 2 - (2 / n_qubits) * Q
    return Q
```

Function that keeps track of the relevant quantities during evolution.

```python
def update_print_evolution(probs, t, spectrum, evolution, state):  
    best_cost = min(spectrum.values())
    best_solutions = {bitstring for bitstring, cost in spectrum.items() if cost==best_cost}
    solution = next(iter(best_solutions))
    try: 
        overlap_ground = probs.get(solution, 0)
    except:
        overlap_ground = None
    average_cost = evaluate_cost(probs, spectrum)
    most_likely_solution = max(probs, key=probs.get)
    overlap_most_lik_sol = probs[most_likely_solution]
    evolution['overlap ground state'].append(overlap_ground)
    evolution['best cost'].append(best_cost)
    evolution['average cost'].append(average_cost)
    evolution['overlap most likely solution'].append(overlap_most_lik_sol)
    evolution['tau'].append(t)

    entanglement = get_entanglement(state)
    evolution['entanglement'].append(entanglement)
        
    string_to_print = 'ov gr = ' + '{:.1e}'.format(overlap_ground) +  ', av cost = ' + '{:.1e}'.format(average_cost) + ', best cost = ' + '{:.1e}'.format(best_cost) + ', ent = ' + '{:.1e}'.format(entanglement) 
                           
    return evolution, string_to_print    
```

Function plotting the evolution of the relevant quantities of a F_VQE instance.

```python
def plot_evolution(evolution):
    instructions_l_axis = {'overlap ground state': {'color': 'green'},
                           'overlap most likely solution': {'color': 'blue', 'linestyle': '--'}}

    instructions_l_axis['entanglement'] = {'color': 'gray'}
        
    instructions_r_axis = {'best cost': {'color': 'red', 'linestyle': '--'},
                           'average cost': {'color': 'red'}}
    fig, ax = plt.subplots()
    fig.patch.set_facecolor('white')  
    plt.title(f'{n_qubits} qubits')
    X = np.arange(n_steps + 1)
    plt.xlim(0, n_steps)
    plt.xlabel('optimisation step')
    
    ax.set_ylabel('overlap and entanglement')
    ax.set_ylim([0, 1])
    for label, style in instructions_l_axis.items():
        Y = np.array(evolution[label])
        if style.get('rescale', False): 
            rescale_factor = max([y for y in Y if not np.isnan(y)]) / ax.get_ylim()[1]
            if rescale_factor > 1:
                rescale_factor = 10 ** np.ceil(np.log10(rescale_factor))
                Y /= rescale_factor
                label += ' x ~' + '{:.0e}'.format(rescale_factor)
            del style['rescale']
        ax.plot(X, Y, **style, label=label)
    
    ax2 = ax.twinx()
    ax.plot([], [], color='red', linewidth=3)
    ax2.set_ylabel('cost', color='tab:' + 'red')
    ax2.tick_params(axis='y', labelcolor='tab:' + 'red')
    for label, style in instructions_r_axis.items():
        ax2.plot(X, np.array(evolution[label]), **style, label=label)
    
    ax.legend(bbox_to_anchor=(1.1, 1), loc='upper left')
    ax2.legend(bbox_to_anchor=(1.1, 0.7), loc='upper left')
    plt.show()
```

# F-VQE algorithm

This is the core of this notebook. The algorithm takes an initial set of parameters and a cost function, iterates n times, and optimizes the circuit so that parameters are returned that give the desired solution with high probability.

```python
def fvqe(params, t, n_steps, spectrum, cost_function,allowPrint=True):
    evolution = {'overlap ground state': [], 'best cost': [], 'average cost': [], 
                 'overlap most likely solution': [], 'tau': [], 'entanglement':[]}    
    for step in range(n_steps):   
        state = get_state(params)
        probs = sample_ansatz(params, state) 
        spectrum = update_spectrum(probs, spectrum,cost_function)
        evolution, string_to_print = update_print_evolution(probs, t, spectrum, evolution, state)  
        if allowPrint:
            print(f"step {step}: " + string_to_print)   
        probs_plus_all, probs_minus_all = sample_shifted_ansatz(params)    
        for i in range(n_params):
            spectrum = update_spectrum(probs_plus_all[i], spectrum,cost_function)
            spectrum = update_spectrum(probs_minus_all[i], spectrum,cost_function)
        
        t, gradient = update_tau(probs_plus_all, probs_minus_all, spectrum)
        params = update_params(params, gradient)
        
    step = n_steps  
    probs = sample_ansatz(params,state)  
    spectrum = update_spectrum(probs, spectrum,cost_function)
    evolution, string_to_print = update_print_evolution(probs, t, spectrum, evolution, state)  
    if allowPrint:
        print(f"step {step}: " + string_to_print)  
    return evolution, params, spectrum
```

```python
t = 1  # initial value of tau.
r = 0.1  # threshold for tau scheduling.
learning_rate = 1
n_steps = 40 # number of F-VQE iterations

spectrum = dict()

n_qubits = 4
n_layers = 2  # number of ansatz circuit layers
n_gates = 3
n_para = n_qubits * n_layers *n_gates  # number of parameters for the ansatz below

initial_params = r_state_1
circuit_params = np.random.rand(n_para)*np.pi/2

params = np.concatenate((initial_params,circuit_params))
n_params = len(params)
```

Finally, we run the experiment and plot the results.

```python
evolution, final_param, spectrum = fvqe(params, t, n_steps,spectrum,cost_function = C1,allowPrint=False)
plot_evolution(evolution)
```

This diagram is full of interesting information. I will discuss all the curves listed in the legend step by step. 
- Starting from the bottom, the best cost is just the cost of the optimal solution. If we use our random state 1, these would be $|0011>$. Logically, it remains constant as we do not vary our cost function.
- The average cost is calculated by taking the cost average of the bit strings, weighted by their probability distribution. For example, if the sample $(0,0,1,1)$ has a probability of $1$ and $C((0,0,1,1))=0$, the average cost is $0$. 
- The entanglement curve shows the evolution of, well, entanglement in the circuit. We see that in the random superposition we start with, we have a strongly entangled state. This is not surprising, since when a state is randomly generated, the probability of generating an entangled state is higher than vice versa. Entanglement decreases over time as our state evolves from a superposition to a particular single state.
- Lastly, overlap most likely solution/ground state give us a measure of $|<\psi|\lambda_0>|^2$

```python
final_state = get_state(final_param)
final_probs = sample_ansatz(final_param,final_state) 
plot_probs(final_probs)
```

Plotting the probability distribution of the states confirms that we have successfully accomplished our task of taking a random input state and returning the state $(0,0,1,1)$.

# Solving for 4 different initial parameters with 

We now extend our code to handle multiple cost functions/ input states.

```python
def C2(x):
    solution = (0,1,0,1)
    return np.sum(np.abs(x-np.array(solution)))
def C3(x):
    solution = (1,0,1,0)
    return np.sum(np.abs(x-np.array(solution)))
def C4(x):
    solution = (1,1,0,0)
    return np.sum(np.abs(x-np.array(solution)))
```

```python
problem_set = [C1,C2,C3,C4]
initial_param_arr = [r_state_1,r_state_2,r_state_3,r_state_4]

def fvqe_on_set(t, n_steps,spectrum,allowPrint=False):
    output_evo = []
    output_params = []
    for i, cost_function in enumerate(problem_set):
        print(f'problem {i+1}')
                
        params = np.concatenate((initial_param_arr[i],np.random.rand(n_para)*np.pi/2))
        spectrum = dict()
        
        evolution, final_params, spectrum = fvqe(params, t, n_steps,spectrum,cost_function=cost_function,allowPrint=False)
        output_evo.append(evolution)
        output_params.append(final_params)
        
    return output_evo,output_params

evo_4, params_4 = fvqe_on_set( t, n_steps,spectrum,allowPrint=False)
```

```python
for final_params in params_4:
    final_state = get_state(final_params)
    final_probs = sample_ansatz(final_params,final_state)
    plot_probs(final_probs)
```
